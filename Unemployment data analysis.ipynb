```python
#Unemployment rate in India (2015 to 2021)

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
url = 'https://raw.githubusercontent.com/HemalathaKoteru/Anudip_Project2/main/unemployment%20analysis11.csv'
df = pd.read_csv(url)

# Filter the data by country name (India)
country_name = 'India'
df_country = df[df['Country Name'] == country_name]

# Extract years and corresponding unemployment rates
years = [int(col.split('-')[1]) for col in df_country.columns if col.startswith('Y-')]
unemployment_rates = [df_country[f'Y-{year}'].iloc[0] for year in years]

# Filter data for the years 2015 to 2021
years_filtered = [year for year in years if 2015 <= year <= 2021]
unemployment_rates_filtered = [unemployment_rates[years.index(year)] for year in years_filtered]

# Plot the line chart
plt.figure(figsize=(10, 6)) # It specifies size of figure
plt.plot(years_filtered, unemployment_rates_filtered, marker='o', linestyle='-')
plt.title(f"Unemployment Rate in {country_name} (2015-2021)")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate")
plt.grid(True)
plt.xticks(years_filtered)
plt.tight_layout()
plt.show()
```

```python
# make pie CHART for Umemployment in india from the year of Y-2001 to Y-2021
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
url = 'https://raw.githubusercontent.com/HemalathaKoteru/Anudip_Project2/main/unemployment%20analysis11.csv'
df = pd.read_csv(url)

# Filter the data by country name (India)
country_name = 'India'
df_country = df[df['Country Name'] == country_name]

# Extract columns for the years 2001 to 2021
years_columns = [f'Y-{year}' for year in range(2001, 2022)]

# Calculate the average unemployment rate for each year
average_unemployment_rates = [df_country[year_column].mean() for year_column in years_columns]

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(average_unemployment_rates, labels=range(2001, 2022), autopct='%1.1f%%', startangle=140)
plt.title(f"Average Unemployment Rate in {country_name} (2001-2021)")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
url = 'https://raw.githubusercontent.com/HemalathaKoteru/Anudip_Project2/main/unemployment%20analysis11.csv'
df = pd.read_csv(url)

# Filter the data for India and China
df_india = df[df['Country Name'] == 'India']
df_china = df[df['Country Name'] == 'China']

# Extract years and population data for India
years = range(2015, 2022)

# Extract population data for India
india_population = [df_india[f'Y-{year}'].iloc[0] for year in years]

# Extract population data for China
china_population = [df_china[f'Y-{year}'].iloc[0] for year in years]

# Plot the bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(years))
plt.bar(index, india_population, bar_width, color='b', label='India')
plt.bar([i + bar_width for i in index], china_population, bar_width, color='r', label='China')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Comparison: India vs China (2015-2021)')
plt.xticks([i + bar_width / 2 for i in index], years)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset containing unemployment rate data
df = pd.read_csv('https://raw.githubusercontent.com/HemalathaKoteru/Anudip_Project2/main/unemployment%20analysis11.csv')

# Filter the DataFrame to get data only for Japan from 2018 to 2021
japan_data = df[df['Country Name'] == 'Japan']

# Extract the columns for years 2018 to 2021
japan_data = japan_data[['Y-2018', 'Y-2019', 'Y-2020', 'Y-2021']]

# Transpose the DataFrame to have years as rows and columns as countries
japan_data = japan_data.transpose()

# Plot a bar chart
plt.figure(figsize=(10, 6))
plt.bar(japan_data.index, japan_data.values.flatten(), color='pink')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate in Japan (2018-2021)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset containing unemployment rate data
df = pd.read_csv('https://raw.githubusercontent.com/HemalathaKoteru/Anudip_Project2/main/unemployment%20analysis11.csv')

# Filter the data to include only the countries with the highest rates in 1991
countries_1991 = ['Lesotho', 'South Africa', 'Djibouti', 'Montenegro', 'North Macedonia']
unemployment_2021_countries_1991 = df[df['Country Name'].isin(countries_1991)]

# Extract unemployment rates for the year 2021
unemployment_2021 = unemployment_2021_countries_1991[['Country Name', 'Y-2021']]

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(unemployment_2021['Country Name'], unemployment_2021['Y-2021'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rates in Selected Countries for the Year 2021')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset containing unemployment rate data
df = pd.read_csv('https://raw.githubusercontent.com/HemalathaKoteru/Anudip_Project2/main/unemployment%20analysis11.csv')

# Specify the countries with the highest unemployment rates in 1991
countries_1991 = ['Lesotho', 'South Africa', 'Djibouti', 'Montenegro', 'North Macedonia']

# Filter the dataset to include only the specified countries for the year 2021
unemployment_2021 = df[df['Country Name'].isin(countries_1991)][['Country Name', 'Y-2021']]

# Find the country with the lowest unemployment rate
lowest_country = unemployment_2021.loc[unemployment_2021['Y-2021'].idxmin()]

# Find the country with the highest unemployment rate
highest_country = unemployment_2021.loc[unemployment_2021['Y-2021'].idxmax()]

# Plotting the bar graph
plt.figure(figsize=(10, 6))
plt.bar(unemployment_2021['Country Name'], unemployment_2021['Y-2021'], color='skyblue')
plt.bar(lowest_country['Country Name'], lowest_country['Y-2021'], color='green', label=f"Lowest: {lowest_country['Country Name']} - {lowest_country['Y-2021']}%")
plt.bar(highest_country['Country Name'], highest_country['Y-2021'], color='red', label=f"Highest: {highest_country['Country Name']} - {highest_country['Y-2021']}%")
plt.xlabel('Country')
plt.ylabel('Unemployment Rate (%)')
plt.title('Comparison of Unemployment Rates in Selected Countries for the Year 2021')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
